```python
import yt_dlp
import openai
import os
from api_senha import senha

# Autenticao
client = openai.OpenAI(api_key = senha)

def baixar_audio_ytdlp(url, caminho_destino, nome_arquivo='audio'):
    os.makedirs(caminho_destino, exist_ok=True)

    caminho_completo = os.path.join(caminho_destino, nome_arquivo + '.mp3')

    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': caminho_completo,
        'quiet': True,
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])
    
    print(f" udio salvo em: {caminho_completo}")
    return caminho_completo

def transcrever_audio_novo(caminho_arquivo):
    with open(caminho_arquivo, "rb") as f:
        transcript = client.audio.transcriptions.create(
            model="whisper-1",
            file=f,
            response_format="text"
        )
    return transcript

#  USO
if __name__ == "__main__":
    link_do_vdeo = 'https://www.youtube.com/shorts/crV6ml-1neM'
    destino = r"C:\Users\guhen\programas\TRANSCREVERDOR-YT"
    nome = "audio_stormer"

    caminho_audio = baixar_audio_ytdlp(link_do_vdeo, destino, nome)
    texto = transcrever_audio_novo(caminho_audio)

    print("\n Transcrio:\n", texto)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
import nltk
from collections import Counter
import string

# Se rodar pela primeira vez:
# nltk.download('stopwords')

from nltk.corpus import stopwords

# Limpeza bsica
texto = texto.lower()
for pontuacao in string.punctuation + "":
    texto = texto.replace(pontuacao, " ")
palavras = texto.split()
stopwords_pt = set(stopwords.words('portuguese'))
palavras_filtradas = [p for p in palavras if p not in stopwords_pt and len(p) > 2]

# Frequncia das palavras
contagem = Counter(palavras_filtradas)
df = pd.DataFrame(contagem.items(), columns=["Palavra", "Frequncia"]).sort_values("Frequncia", ascending=False)

print("Top 20 palavras mais frequentes:\n", df.head(20))

# Grfico de barras
plt.figure(figsize=(12, 5))
df.head(20).plot.bar(x="Palavra", y="Frequncia", color="#1d4851", legend=False)
plt.title("Top 20 Palavras Mais Frequentes")
plt.xlabel("Palavra")
plt.ylabel("Frequncia")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Nuvem de palavras
wordcloud = WordCloud(
    width=800, height=400,
    background_color="black",
    colormap="cool",
    stopwords=stopwords_pt
).generate(" ".join(palavras_filtradas))

plt.figure(figsize=(15, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Nuvem de Palavras")
plt.show()

# Grfico de pizza das top 10
df_top10 = df.head(10)
plt.figure(figsize=(8, 8))
plt.pie(df_top10["Frequncia"], labels=df_top10["Palavra"], autopct='%1.1f%%', startangle=140, colors=plt.cm.cool(np.linspace(0,1,10)))
plt.title("Top 10 Palavras Mais Frequentes")
plt.show()

# Histograma log da frequncia
plt.figure(figsize=(10, 4))
plt.hist(df["Frequncia"], bins=30, color="#1d4851", edgecolor="white", log=True)
plt.title("Histograma da Frequncia das Palavras (Escala Logartmica)")
plt.xlabel("Frequncia")
plt.ylabel("Nmero de Palavras (log)")
plt.tight_layout()
plt.show()

# Palavras mais longas/curtas
df["Tamanho"] = df["Palavra"].apply(len)
print("\nPalavras mais longas:")
print(df.nlargest(5, "Tamanho")[["Palavra", "Tamanho", "Frequncia"]])
print("\nPalavras mais curtas:")
print(df.nsmallest(5, "Tamanho")[["Palavra", "Tamanho", "Frequncia"]])

# Estatsticas bsicas
media = np.mean(df["Frequncia"])
mediana = np.median(df["Frequncia"])
moda_valor = df["Frequncia"].mode()[0]
palavra_media = df.iloc[(df["Frequncia"] - media).abs().idxmin()]["Palavra"]
palavra_mediana = df.iloc[(df["Frequncia"] - mediana).abs().idxmin()]["Palavra"]
palavras_moda = df[df["Frequncia"] == moda_valor]["Palavra"].tolist()

print(f"\n Mdia da Frequncia: {media:.2f} (Palavra mais prxima: {palavra_media})")
print(f" Mediana: {mediana} (Palavra mais prxima: {palavra_mediana})")
print(f" Moda: {moda_valor} (Palavras: {', '.join(palavras_moda)})")

# Bigramas mais frequentes
from itertools import tee, islice

def bigramas(lista):
    a, b = tee(lista)
    next(b, None)
    return zip(a, b)

bigramas_mais_frequentes = Counter(bigramas(palavras_filtradas))
df_bigramas = pd.DataFrame(bigramas_mais_frequentes.most_common(10), columns=["Bigramas", "Frequncia"])
print("\nTop 10 bigramas mais frequentes:\n", df_bigramas)

# Insights automticos
palavra_mais_frequente = df.iloc[0]["Palavra"]
print(f"""
 Insights:
 Palavra mais frequente: {palavra_mais_frequente}
 Total de palavras nicas: {df.shape[0]}
 Frequncia mdia: {media:.2f}
 A nuvem de palavras destaca os principais termos visualmente.
""")
```
